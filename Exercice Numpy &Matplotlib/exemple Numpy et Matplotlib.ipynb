#courbes multiples dans une même figure

```python
plt.figure(figsize=(8,4))
x=np.linspace(0,10)
plt.plot(x,np.sin(x))
plt.plot(x,np.sin(x),color='blue',linestyle='solid',label='sin(x)')
plt.plot(x,np.sin(x-1),color='g',linestyle='dashed',label='sin(x-1)')
plt.plot(x,np.sin(x-2),color='0.75',linestyle='dashdot',label='sin(x-2)')
plt.plot(x,np.sin(x-3),color='#FF0000',linestyle='dotted',label='sin(x-3)')
plt.title("un exemple de graphe")
plt.legend(loc='lower right')
```

***Méthodes statistiques de Nympy***


```python

arr = np.array([[1,5,6],[4,7,2],[3,1,9]])
print("le plus grand element dans le tableau :",np.max(arr))
print("le plus maximaux par colonne:",np.max(arr,axis=0))
print("le plus maximaux par ligne :",np.max(arr,axis=1))
print("le plus minimaux par rangée :",np.min(arr,axis=1))
print("*************************")
print("somme de touts les éléments:",np.sum(arr))
print("somme cumulative:\n",np.cumsum(arr))
print("somme cumulative par colonne :",np.cumsum(arr ,axis=0))
print("somme cumulative par ligne :",np.cumsum(arr ,axis=1))
print("*************************")
print("la moyen des éléments par colonne :",np.around(np.mean(arr ,axis=0), 3))
print("la moyen des éléments par ligne:",np.around(np.mean(arr ,axis=1),3))
print("la moyen des éléments :",np.around(np.mean(arr),3))
print("*************************")
print("la médian des éléments :",np.median(arr))
print("la médian des éléments par colonne :",np.median(arr,axis=0))
print("la médian des éléments  par ligne:",np.median(arr ,axis=1))
print("*************************")
print("l'ecartype des éléments :",np.around(np.std(arr),3))
print("*************************")
print("la variance des éléments :",np.around(np.var(arr),3))
```

#tableau avec des valeurs nulles

```python
print("le plus grand element dans le tableau :",np.nanmax(arr))
print("le plus maximaux par colonne:",np.nanmax(arr,axis=0))
print("le plus maximaux par ligne :",np.nanmax(arr,axis=1))
print("le plus minimaux par rangée :",np.nanmin(arr,axis=1))
print("somme de touts les éléments:",np.nansum(arr))
print("somme cumulative:\n",np.nancumsum(arr))
print("somme cumulative le long de chaque  colonne :",np.nancumsum(arr ,axis=0))
```

***Créer un histogramme avec numpy***



```python

x=np.random.randint(15,60,size=(27))
x[7]=-10
x[12]=-10
print("données:",x)
counts,bins=np.histogram(x,np.arange(-10,70,10))
print('count',counts,'\n bins',bins)
plt.subplots()
plt.hist(bins[:-1],bins ,weights=counts,rwidth=0.9,color="b",alpha=0.5,label="Deli de facturation")
plt.xlabel("Delai de facturation")
plt.ylabel("Nombre de clients")
plt.show()
```
